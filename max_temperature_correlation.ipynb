```python
# Max Temperature Correlation between Miami, New York and Los Angeles


# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns


# Read Miami csv file
miami = pd.read_csv("/content/Miami.csv")

# Split the DATE column into two columns YEAR and MONTH
miami[['YEAR','MONTH']] = miami.DATE.str.split("-",expand=True)
miami['YEAR'] = miami['YEAR'].astype(int)

# Create dataframe only with values of the considered time period
miami_1951_2020 = miami[miami['YEAR'] >= 1951]
miami_1951_2020 = miami_1951_2020[miami_1951_2020['YEAR'] <= 2020]



# Read New York csv file
nyc = pd.read_csv("/content/NY_JFK.csv")

# Split the DATE column into two columns YEAR and MONTH
nyc[['YEAR','MONTH']] = nyc.DATE.str.split("-",expand=True)
nyc['YEAR'] = nyc['YEAR'].astype(int)

# Create dataframe only with values of the considered time period
nyc_1951_2020 = nyc[nyc['YEAR'] >= 1951]
nyc_1951_2020 = nyc_1951_2020[nyc_1951_2020['YEAR'] <= 2020]




# Read Los Angeles csv file
lax = pd.read_csv("/content/LAX.csv")

# Split the DATE column into two columns YEAR and MONTH
lax[['YEAR','MONTH']] = lax.DATE.str.split("-",expand=True)
lax['YEAR'] = lax['YEAR'].astype(int)

# Create dataframe only with values of the considered time period
lax_1951_2020 = lax[lax['YEAR'] >= 1951]
lax_1951_2020 = lax_1951_2020[lax_1951_2020['YEAR'] <= 2020]


# Perform Pearson R-test for all city combinations
print("Pearson R-test comparing Miami and New York:",st.pearsonr(miami_1951_2020["TMAX"],nyc_1951_2020["TMAX"]))
print("Pearson R-test comparing Miami and Los Angeles:",st.pearsonr(miami_1951_2020["TMAX"],lax_1951_2020["TMAX"]))
print("Pearson R-test comparing Los Angeles and New York:",st.pearsonr(lax_1951_2020["TMAX"],nyc_1951_2020["TMAX"]))


# Create regression plots for all city combinations

# To see the image for one specific city, leave the code for the remaining to cities as a comment

#mia_nyc = sns.regplot(x=miami_1951_2020["TMAX"], y=nyc_1951_2020["TMAX"])
#mia_nyc.set(xlabel='Miami Max Temp (C)', ylabel='New York Max Temp (C)')

#mia_lax = sns.regplot(x=miami_1951_2020["TMAX"], y=lax_1951_2020["TMAX"])
#mia_lax.set(xlabel='Miami Max Temp (C)', ylabel='Los Angeles Max Temp (C)')

lax_nyc = sns.regplot(x=lax_1951_2020["TMAX"], y=nyc_1951_2020["TMAX"])
lax_nyc.set(xlabel='Los Angeles Max Temp (C)', ylabel='New York Max Temp (C)')
```
