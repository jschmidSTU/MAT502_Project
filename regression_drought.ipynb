```python
# Regression: Drought Prediction


# Import necessary libraries
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm


# Read drought csv file and rename 3 columns to make them more understandable
drought_df = pd.read_csv("/content/drought.csv")
drought_df.rename(columns={"PRECTOT": "precipitation", "PS": "surface_pressure", "QV2M": "humidity"}, inplace=True)


# Eliminate all rows where the value for score is empty
drought_df['score'].replace('', np.nan, inplace=True)
drought_df.dropna(subset=['score'], inplace=True)
print(drought_df)


# Linear Regression: First approach using all parameters
model = smf.ols('score ~ fips + date + precipitation + surface_pressure + humidity + T2M + T2MDEW + T2MWET + T2M_MAX + T2M_MIN + T2M_RANGE + TS + WS10M + WS10M_MAX + WS10M_MIN + WS10M_RANGE + WS50M + WS50M_MAX + WS50M_MIN + WS50M_RANGE', drought_df).fit()
print(model.summary())
print(sm.stats.anova_lm(model, typ = 2))


# Drop all parameters with p-value > 0.05 = significance level alpha
model = smf.ols('score ~ fips + date + precipitation + surface_pressure + humidity + T2M + T2MDEW + T2MWET + TS + WS10M', drought_df).fit()
print(model.summary())
print(sm.stats.anova_lm(model, typ = 2))
```
